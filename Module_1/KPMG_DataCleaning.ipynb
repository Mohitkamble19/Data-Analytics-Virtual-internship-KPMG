```python
df2['gender'] = df2['gender'].astype('category')
```

```python
df2["gender"].replace({'F' : 'Female', 'Femal' : 'Female','M':'Male', 'U':'Unspecified'}, inplace=True)
```

```python
df2["gender"].value_counts()
```

```python
df2["DOB"].value_counts(
    
)
```

```python
df2['DOB']=df2['DOB'].fillna(method='ffill')
```

```python
df2['job_title'].value_counts()
```

```python
df2['job_title']= df2['job_title'].fillna(method='bfill')
```

```python
df2['job_industry_category'].value_counts()
```

```python
df2['job_industry_category']= df2['job_industry_category'].fillna(method='ffill')
```

```python
# convert to boolean data type
f = {'N': False, 'Y': True}
df2['deceased_indicator'] = df2['deceased_indicator'].map(f).fillna(df2['deceased_indicator'])
```

```python
# convert to boolean data type
g = {'No': False, 'Yes': True}
df2['owns_car'] = df2['owns_car'].map(g).fillna(df2['owns_car'])
```

```python
mean=np.nanmedian(df2['tenure'])
```

```python
df2['tenure'].fillna(mean, inplace = True)
```

```python
# make sure that the column names are in the correct format
df2.columns = map(str.lower, df2.columns)
df2.columns = map(str.strip, df2.columns)
```

```python
df2.info()
```

```python
df2.to_csv('./data2_cleaned.csv',index=False)
```

```python
df4['last_name'] = df4.groupby('first_name').last_name.bfill().ffill()
```

```python
df4['state'].value_counts()
```

```python
df4['country'].value_counts()
```

```python
df4['address'].nunique()
```

```python
# make sure all dates are in the correct format
df4['DOB'] = df4['DOB'].dt.date
df4['DOB']
```

```python
# fill the missing date of birth with a random choice of date from the existing datapoints
df4['DOB'].fillna(lambda x: np.random.choice(df4['DOB']), inplace=True)
```

```python
# check the frequency of different job titles
df4['job_title'].value_counts()
```

```python

df4['job_title']= df4['job_title'].fillna(method='bfill')
```

```python
df4['job_industry_category'].value_counts()
```

```python
df4['job_industry_category']= df4['job_industry_category'].fillna(method='ffill')
```

```python
# make sure that the column names are in the correct format
df4.columns = map(str.lower, df4.columns)
df4.columns = map(str.strip, df4.columns)
```

```python

df4['value'] = df4['value'].round(3)
df4
```
